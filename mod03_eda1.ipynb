### On the training set, create pairwise scatterplots for each of these 6 columns with the 'Salary' variable.

```python
x = np.array(Hitters.Salary)
y = np.array(Hitters.Runs)
plt.scatter(x, y)
plt.xlabel("Salary")
plt.ylabel("Runs")
plt.title("Salary vs Runs")
plt.show()

x = np.array(Hitters.Salary)
y = np.array(Hitters.AtBat)
plt.scatter(x, y)
plt.xlabel("Salary")
plt.ylabel("Runs")
plt.title("Salary vs At Bats")
plt.show()

x = np.array(Hitters.Salary)
y = np.array(Hitters.Hits)
plt.scatter(x, y)
plt.xlabel("Salary")
plt.ylabel("Runs")
plt.title("Salary vs Hits")
plt.show()

x = np.array(Hitters.Salary)
y = np.array(Hitters.HmRun)
plt.scatter(x, y)
plt.xlabel("Salary")
plt.ylabel("Runs")
plt.title("Salary vs Home Runs")
plt.show()

x = np.array(Hitters.Salary)
y = np.array(Hitters.RBI)
plt.scatter(x, y)
plt.xlabel("Salary")
plt.ylabel("Runs")
plt.title("Salary vs RBI")
plt.show()

x = np.array(Hitters.Salary)
y = np.array(Hitters.Walks)
plt.scatter(x, y)
plt.xlabel("Salary")
plt.ylabel("Runs")
plt.title("Salary vs Walks")
plt.show()


```
